### Problem 1

Consider a digital clock that shows hours and minutes. The day start at 00:00 and ends at 24:00. Assume
that the four digits shown in the clock actually is one single number. For example 13:15 is to be interpreted
as 1315. How many prime numbers are shown by the digital clock in a day?

```r
#creating time range i.e.,from 0000-2359

HourTick <-(sprintf("%01d",0:24))
MinTick<-(sprintf("%02d",0:59))

k=1

a <-c()
for(i in 1:24){
  for(j in 0:59){
    j=j+1
    a[k]<-((paste0(HourTick[i],MinTick[j],sep=" ")))
    k=k+1
  }
  i=i+1
}

a<-as.numeric(a)
a
```

```r
#Now here we go !!
#creating functions

TheTimeIs <- function(a)
{
  for (i in 2:(a-1))
  {
    if (a %% i == 0)
    {
      return (FALSE)
    }
  }
  return (TRUE)
}

clockPrimes <- function(time){
  counter <- 1
 
  for (i in 2:time){
    if (TheTimeIs(i) == TRUE){
      counter <- counter + 1
    }
  }
  return (counter)
}

cat("No of Prime Numbers in this:")
clockPrimes(111)
```
